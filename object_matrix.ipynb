```python
import os
import pandas as pd
import numpy as np
import sklearn
import csv
```

```python
df = pd.read_csv('Charades_vu17_train.csv')
```

```python
df.head()
```

```python
print(df.shape)
df = df[pd.notnull(df['objects'])] # drop NA in column['actions']
print(df.shape)
df['origin_index'] = df.index.values
df = df.reset_index(drop=True)
df.head()
```

```python
dat_id = df['id']
dat_actions = df['objects']
dat_origin_index = df['origin_index']
dat = pd.concat([dat_id, dat_origin_index, dat_actions], axis=1)
split_arr = dat['objects'].str.split(';')
```

```python
dat = dat.drop('objects', 1)
```

```python
dat['split'] = split_arr
```

```python
mydict = dict()
for objects in dat['split']:
    for obj in objects:
        if(mydict.get(obj) == None):
            mydict[obj] = 1
        else:
            mydict[obj] += 1
```
