```python
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dat = open('log_fc_rgb.txt', 'r', encoding='UTF-8')
arr_loss = np.zeros(1244)
arr_acc = np.zeros(1244)

val_loss = np.zeros(1243)
val_rank1 = np.zeros(1243)
val_rank5 = np.zeros(1243)

i = 0
j = 0
for line in dat.readlines():
    arr = line.split(' ')
    if arr[0] == "TRAIN":
        loss = arr[2]
        acc = arr[3]
        arr_loss[i] = loss
        arr_acc[i] = acc
        i += 1
    elif arr[0] == "VAL":
        loss = arr[2]
        r1 = arr[3]
        r5 = arr[4]
        val_loss[j] = loss
        val_rank1[j] = r1
        val_rank5[j] = r5
        j += 1
```

```python
plt.plot(arr_loss) # only one descending 
plt.plot(val_loss)
plt.plot(val_rank1)
plt.plot(val_rank5)
```
